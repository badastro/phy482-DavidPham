```python
e = np.arange(0,10,0.01)
r = 10*np.exp(-e)

plt.figure(figsize=(12,6))
plt.plot(e,r)
plt.xlabel('r')
plt.ylabel('E(r)')
```

```python
epsilon = 8.85e-12
rho = 1e-5
R = 1
```

```python
rin = np.arange(0,R,0.01)
rout = np.arange(R,10*R,0.01)
```

```python
Ein = rho*rin/3/epsilon
Eout = rho*R**3/rout/3/epsilon
```

```python
plt.figure(figsize=(8,8))
plt.plot(rin,Ein)
plt.plot(rout,Eout)
plt.xlabel('r')
plt.ylabel('E(r)')
```
